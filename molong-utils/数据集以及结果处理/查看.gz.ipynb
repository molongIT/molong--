```python
import os
import nibabel as nib
import matplotlib.pyplot as plt

def display_image(img):
    """显示给定的NIfTI图像"""
    # 获取图像数据
    data = img.get_fdata()
    print(data.shape)
    # 确保图像数据是二维的
    if data.ndim == 2:
        plt.imshow(data, cmap='gray')
        plt.axis('off')
        plt.title('NIfTI Image')
        plt.show()
    else:
        print("图像数据不是二维的，无法显示")

def traverse_and_display(directory,num):
    """遍历目录下的所有.nii.gz文件并显示它们"""
    for root, dirs, files in os.walk(directory):
        for file in files:
            if(num < 0):
                return
            num = num - 1
            if file.endswith('.nii.gz'):
                file_path = os.path.join(root, file)
                print(f"Processing file: {file_path}")
                img = nib.load(file_path)
                display_image(img)

num = 3
```

```python

# 指定需要遍历的目录
directory = '/home/pxl/myProject/血管分割/DSCNet/DSCNet_2D_opensource/Results/DRIVE/DSCNet'

# 调用函数遍历并显示
traverse_and_display(directory,num)
```
